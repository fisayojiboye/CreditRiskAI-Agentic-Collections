### Level 3: Full LLM-Powered Agentic AI
In Level 3, we will upgrade the agent to use real LLM reasoning (step-by-step thinking), tools, and more natural decision-making.

```python
import pandas as pd
import json
import os
from datetime import datetime

print("✅ Level 3 - LLM-Powered Agentic AI Initialized")
```

##### 2: Load Data & Memory

```python
# Load the final dataset
df = pd.read_excel("C:/Users\HP\Projects\Gen AI Powered Data Analytics\Final_Dataset_with_Agent_Level2.xlsx")

# Load existing memory
memory_file = 'logs/customer_memory.json'
if os.path.exists(memory_file):
    with open(memory_file, 'r') as f:
        customer_memory = json.load(f)
    print(f"✅ Loaded memory for {len(customer_memory)} customers")
else:
    customer_memory = {}
    print("ℹ️ Starting with fresh memory")

print(f"Total customers: {len(df)}")
```

##### 3.5: Risk Scoring Function

```python
# Code Block 3.5: Risk Scoring Function (Required for Level 3)
def calculate_risk_score(row):
    """Calculate risk score (0-100) based on key features"""
    score = 0
    
    # Missed Payments - Strongest factor
    if row['Missed_Payments'] >= 4:
        score += 50
    elif row['Missed_Payments'] >= 2:
        score += 25
    
    # Credit Utilization
    if row['Credit_Utilization'] > 0.70:
        score += 30
    elif row['Credit_Utilization'] > 0.50:
        score += 15
    
    # Credit Score (lower = higher risk)
    if row['Credit_Score'] < 500:
        score += 15
    elif row['Credit_Score'] < 600:
        score += 8
    
    return min(score, 100)
```

### 3: LLM-Powered Reasoning Function (The Brain)

```python
def llm_reasoning(customer_row, memory):
    """
    Level 3: Uses step-by-step LLM-style reasoning.
    """
    cust_id = customer_row['Customer_ID']
    history = memory.get(cust_id, {"interactions": []})
    prev_count = len(history["interactions"])
    
    # Build context for reasoning
    context = f"""
    Customer Profile:
    - Missed Payments: {customer_row['Missed_Payments']}
    - Credit Utilization: {customer_row['Credit_Utilization']:.2f}
    - Credit Score: {customer_row['Credit_Score']}
    - Age: {customer_row['Age']}
    - Employment: {customer_row['Employment_Status']}
    - Previous Interactions: {prev_count}
    """
    
    # Simulate advanced reasoning (we will connect real LLM later)
    risk_score = calculate_risk_score(customer_row)
    
    if risk_score >= 80:
        decision = "ESCALATE"
        reason = "Critical risk level. Multiple red flags present. Immediate human intervention required."
    elif risk_score >= 60:
        decision = "PAYMENT_PLAN"
        reason = "High risk with concerning utilization and missed payments. Structured support needed."
    elif risk_score >= 35:
        decision = "SMS_REMINDER"
        reason = "Moderate risk. Early intervention with reminder may prevent escalation."
    else:
        decision = "MONITOR"
        reason = "Low immediate risk. Continue passive monitoring."
    
    # Record interaction
    history["interactions"].append({
        "timestamp": datetime.now().strftime("%Y-%m-%d %H:%M"),
        "risk_score": risk_score,
        "action": decision,
        "reason": reason
    })
    
    memory[cust_id] = history
    
    return decision, reason, risk_score, context
```

##### 4: Enhanced Guardrails

```python
# Code Block 4: Enhanced Guardrails for Level 3
def apply_guardrails_level3(customer_row, decision, context):
    """
    More advanced guardrails with reasoning context.
    """
    age = customer_row.get('Age', 30)
    employment = str(customer_row.get('Employment_Status', '')).lower()
    missed = customer_row.get('Missed_Payments', 0)
    
    # Guardrail 1: Protect vulnerable customers
    if age >= 65 or 'retired' in employment or 'unemployed' in employment:
        if decision == "ESCALATE":
            return "PAYMENT_PLAN", "Softened from ESCALATE due to vulnerable customer status"
        elif decision == "PAYMENT_PLAN":
            return "SMS_REMINDER", "Further softened for retired/unemployed customer"
    
    # Guardrail 2: First-time or low missed payments
    if missed <= 1 and decision in ["ESCALATE", "PAYMENT_PLAN"]:
        return "SMS_REMINDER", "Softened action - customer has very few missed payments"
    
    # Guardrail 3: Very young customers
    if age < 22 and decision == "ESCALATE":
        return "PAYMENT_PLAN", "Young customer - escalated action softened"
    
    return decision, "No guardrail triggered - action approved"
```

##### 5: Main Level 3 Agent Runner

```python
# Code Block 5: Main Level 3 Agent Runner
def run_level3_agent(customer_row):
    """
    Full Level 3 Agent: Combines LLM-style reasoning + advanced guardrails + memory.
    """
    cust_id = customer_row['Customer_ID']
    
    # Step 1: Advanced Reasoning
    decision, reason, risk_score, context = llm_reasoning(customer_row, customer_memory)
    
    # Step 2: Apply Enhanced Guardrails
    final_action, guardrail_reason = apply_guardrails_level3(customer_row, decision, context)
    
    # Step 3: Create detailed log
    log = {
        'Customer_ID': cust_id,
        'Risk_Score': risk_score,
        'Initial_Decision': decision,
        'Final_Action': final_action,
        'Reason': reason,
        'Guardrail_Applied': guardrail_reason,
        'Timestamp': datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    }
    
    return log, final_action
```

##### 6: Test Level 3 Agent

```python
# Code Block 6: Test Level 3 Agent
print("🚀 Testing Level 3 LLM-Powered Agentic Collections Agent\n")

# Test on 6 customers
test_customers = df.sample(6, random_state=42)

for idx, customer in test_customers.iterrows():
    log, final_action = run_level3_agent(customer)
    
    print(f"Customer: {customer['Customer_ID']}")
    print(f"Risk Score : {log['Risk_Score']}")
    print(f"Initial Decision: {log['Initial_Decision']}")
    print(f"Final Action   : {final_action}")
    print(f"Reason         : {log['Reason']}")
    if log['Guardrail_Applied'] != "No guardrail triggered - action approved":
        print(f"Guardrail      : {log['Guardrail_Applied']}")
    print("-" * 70)
```
