### Level 2: Full Agentic AI with Memory & Reasoning
##### What Level 2 will add:

Memory: The agent will remember previous interactions with each customer.
Better Reasoning: Instead of hard-coded rules only, it will use step-by-step reasoning (more "agentic").
State Management: Track conversation history per customer.
Dynamic Decision Making: The agent can change strategy based on past responses.

This is where the agent starts feeling truly intelligent

```python
# LEVEL 2 - Agentic AI Collections Agent with Memory
# Upgrading from Level 1

import pandas as pd
import numpy as np
from datetime import datetime
import json
import os
from datetime import datetime

# Create memory storage (simple dictionary for now)
customer_memory = {}

print("✅ Level 2 Agentic AI Collections Agent Initialized")
```

---
##### 2: Load Previous Data & Memory

```python
# Load the dataset (same as before)
df = pd.read_excel('/Users\HP\Projects\Gen AI Powered Data Analytics\Cleaned_Delinquency_Dataset_with_New_Target.xlsx')

# Load existing memory if available
memory_file = 'logs/customer_memory.json'
if os.path.exists(memory_file):
    with open(memory_file, 'r') as f:
        customer_memory = json.load(f)
    print(f"✅ Loaded existing memory for {len(customer_memory)} customers")
else:
    print("ℹ️ No previous memory found. Starting fresh.")

print(f"Total customers loaded: {len(df)}")
```

##### 3: Enhanced Risk Scoring (Same as Level 1 but reusable)

```python
def calculate_risk_score(row):
    score = 0
    if row['Missed_Payments'] >= 4:
        score += 50
    elif row['Missed_Payments'] >= 2:
        score += 25
    
    if row['Credit_Utilization'] > 0.70:
        score += 30
    elif row['Credit_Utilization'] > 0.50:
        score += 15
    
    if row['Credit_Score'] < 500:
        score += 15
    elif row['Credit_Score'] < 600:
        score += 8
    
    return min(score, 100)
```

##### 4: Agentic Reasoning + Decision Making (The Brain)

```python
def agentic_reasoning(customer_row, memory):
    """
    Level 2 Agent: Makes decisions with reasoning and memory.
    """
    cust_id = customer_row['Customer_ID']
    risk_score = calculate_risk_score(customer_row)
    
    # Get customer's history
    history = memory.get(cust_id, {"interactions": []})
    prev_interactions = len(history["interactions"])
    
    # Reasoning logic
    if risk_score >= 75 or customer_row['Missed_Payments'] >= 5:
        decision = "ESCALATE"
        reason = "Very high risk - immediate human intervention required"
    elif risk_score >= 50 or customer_row['Missed_Payments'] >= 4:
        decision = "PAYMENT_PLAN"
        reason = "High risk with repeated missed payments - recommend structured plan"
    elif risk_score >= 30 or customer_row['Missed_Payments'] >= 2:
        decision = "SMS_REMINDER"
        reason = "Moderate risk - start with automated reminder"
    else:
        decision = "MONITOR"
        reason = "Low risk - continue monitoring"
    
    # Record this interaction in memory
    history["interactions"].append({
        "timestamp": datetime.now().strftime("%Y-%m-%d %H:%M"),
        "risk_score": risk_score,
        "action": decision,
        "reason": reason
    })
    
    memory[cust_id] = history
    
    return decision, reason, risk_score, prev_interactions
```

##### 6: Guardrails Function

```python
# Code Block 6: Guardrails (Ethical & Fairness Checks)
def apply_guardrails(customer_row, decision):
    """
    Applies ethical guardrails before final action.
    """
    age = customer_row['Age']
    employment = str(customer_row['Employment_Status']).lower()
    missed = customer_row['Missed_Payments']
    
    # Guardrail 1: Protect vulnerable customers
    if age >= 65 or 'retired' in employment or 'unemployed' in employment:
        if decision == "ESCALATE":
            return "PAYMENT_PLAN"   # Soften action
        elif decision == "PAYMENT_PLAN":
            return "SMS_REMINDER"
    
    # Guardrail 2: First-time mild offenders
    if missed <= 1 and decision != "MONITOR":
        return "SMS_REMINDER"
    
    # Guardrail 3: Very young customers (under 22)
    if age < 22 and decision == "ESCALATE":
        return "PAYMENT_PLAN"
    
    return decision
```

##### 7: Main Agent Runner (The Full Agent)

```python
# Code Block 7: Main Agent Function (Level 2)
def run_collections_agent(customer_row):
    """
    Full Level 2 Agent: Combines reasoning + memory + guardrails.
    """
    cust_id = customer_row['Customer_ID']
    
    # Step 1: Reasoning
    decision, reason, risk_score, prev_count = agentic_reasoning(customer_row, customer_memory)
    
    # Step 2: Apply Guardrails
    final_action = apply_guardrails(customer_row, decision)
    
    # Step 3: Create log
    log = {
        'Customer_ID': cust_id,
        'Risk_Score': risk_score,
        'Initial_Decision': decision,
        'Final_Action': final_action,
        'Reason': reason,
        'Previous_Interactions': prev_count,
        'Timestamp': datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    }
    
    return log, final_action
```

##### 8: Test the Level 2 Agent

```python
# Code Block 8: Test Level 2 Agent
print("🚀 Testing Level 2 Agentic Collections Agent\n")

# Test on 8 random customers
test_customers = df.sample(8, random_state=42)

results = []

for idx, customer in test_customers.iterrows():
    log, final_action = run_collections_agent(customer)
    results.append(log)
    
    print(f"Customer {customer['Customer_ID']:10} | "
          f"Risk: {log['Risk_Score']:3} | "
          f"Action: {final_action:12} | "
          f"Prev Interactions: {log['Previous_Interactions']}")

print("\n✅ Level 2 Agent testing completed!")
```

##### 9: Save Memory & Logs

```python
# Code Block 9: Save Memory and Logs
# Save customer memory
os.makedirs('logs', exist_ok=True)

with open('logs/customer_memory.json', 'w') as f:
    json.dump(customer_memory, f, indent=2)

print("✅ Customer memory saved successfully")
print(f"Total customers in memory: {len(customer_memory)}")

# Optional: Save all logs to CSV
logs_df = pd.DataFrame(results)
logs_df.to_csv('logs/agent_decisions_log.csv', index=False)
print("✅ Agent decisions logged to CSV")
```
